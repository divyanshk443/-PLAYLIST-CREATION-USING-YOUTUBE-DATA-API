```python
from apiclient.discovery import build
import argparse
import csv
```

```python
api_key ="Api-key" #told not to submit the api key
DEVELOPER_KEY = api_key
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"
```

```python
youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=DEVELOPER_KEY)
```

## Query 1: 
## Api used=youtube.search.list

```python
query="computer network"
 # calling the search.list method to 
  # retrieve youtube search results 
search_keyword = youtube.search().list(q = query, part = "id, snippet",maxResults = 50).execute() 
       
    # extracting the results from search response 
results = search_keyword.get("items", []) 
   
    # empty list to store video,  
    # channel, playlist metadata 
a=[]
b=[]
video_title=[]
video_id=[]
next_page_token = None
while 1:
    a=youtube.search().list(q=query,part='id,snippet',
                            maxResults=50,
                            pageToken=next_page_token).execute() 
    b=b+a['items']
    next_page_token = a.get('nextPageToken')
    if next_page_token is None:
            break
   
results=b

    # extracting required info from each result object
for result in results: 
            
        # video result object 
        if result['id']['kind'] == "youtube#video": 
            video_title.append("% s" % (result["snippet"]["title"])) 
            video_id.append("% s" % (result["id"]["videoId"])) 
            
```

## Api used=youtube.video.list
####  For finding statistics and published_at column

```python
from datetime import datetime
TimeQueried=[]
publishedAt=[]
statistics=[]
for i in range(len(video_id)):
   list_videos_byid = youtube.videos().list(id = video_id[i], 
          part = "id, snippet, contentDetails, statistics").execute() 

 # extracting the results
   results = list_videos_byid.get("items", []) 
  
     
   for result in results: 
      TimeQueried.append(datetime.now().strftime('%H:%M:%S:%f')[:-3])
      publishedAt.append("(% s)" % (result["snippet"]["publishedAt"])) 
      statistics.append("(% s)" % ( result["statistics"])) 
      
```

### For finding the duration part 

```python
contentDetails=[]
for i in range(len(video_id)):
   list_videos_byid = youtube.videos().list(id = video_id[i], 
          part = "id, snippet, contentDetails").execute() 

 # extracting the results from search response 
   results = list_videos_byid.get("items", []) 
  
     
   for result in results: 
      contentDetails.append('(%s)' % result['contentDetails'])
```

### converting statistics part into a list of dictionary  

```python
ans=[]
for i in range(len(statistics)):
  test_str = statistics[i]
  
# printing original string 
 
# Convert key-value String to dictionary 
# Using map() + split() + loop 
  res = [] 
  for sub in test_str.split(', '): 
     if ':' in sub: 
        res.append(map(str.strip, sub.split(':', 1))) 

  res = dict(res) 
  ans.append(res)
ans
```

### converting contentdetails part into a list of dictionary  

```python
ans2=[]
for i in range(len(contentDetails)):
  test_str = contentDetails[i]
  
# printing original string 
 
# Convert key-value String to dictionary 
# Using map() + split() + loop 
  res = [] 
  for sub in test_str.split(', '): 
     if ':' in sub: 
        res.append(map(str.strip, sub.split(':', 1))) 

  res = dict(res) 
  ans2.append(res)
```

### creating the dataframe for query 1 part

```python
import pandas as pd
df=pd.DataFrame(ans)
dfs=pd.DataFrame(ans2)
df['duration']=dfs["({'duration'"]
df['Topic'] = pd.Series(['computer network' for x in range(len(df.index))])
df['TimeQueried']=pd.DataFrame(TimeQueried)
df['publishedAt']=pd.DataFrame(publishedAt)
df['video_id']=video_id
df['video_title']=video_title
```

## Data manipulation and reordering the columns

```python
df_final['viewcount']=df_final["({'viewCount'"]
df_final['likeCount']=df_final["'likeCount'"]
df_final['dislikeCount']=df_final["'dislikeCount'"]
df_final['favoriteCount']=df_final["'favoriteCount'"]
df_final = df_final[['Topic', 'TimeQueried', 'video_title', 'video_id','publishedAt','duration','viewcount','likeCount','dislikeCount','favoriteCount']]
df_final
```

```python
# for ignoring the warnings
pd.options.mode.chained_assignment = None  
```

## Data analysis and Manipulation

```python
# deleting the first and last character of viewcount,likecount,dislikecount,favoritecount and duration as this is colon
df_final['viewcount']=df_final['viewcount'].map(lambda x: str(x)[1:-1])
df_final['likeCount']=df_final['likeCount'].map(lambda x: str(x)[1:-1])
df_final['dislikeCount']=df_final['dislikeCount'].map(lambda x: str(x)[1:-1])
df_final['favoriteCount']=df_final['favoriteCount'].map(lambda x: str(x)[1:-1])
df_final['duration']=df_final['duration'].map(lambda x: str(x)[1:-1])

# replacing the unwanted string 'a' in viewcount and likecount with 0
df_final["viewcount"].replace({"a": "0"}, inplace=True)
df_final["likeCount"].replace({"a": "0"}, inplace=True)
df_final['viewcount'] = pd.to_numeric(df_final['viewcount'])
df_final['likeCount'] = pd.to_numeric(df_final['likeCount'])
# for this map lambda part i have used stackoverflow
```

```python
df_final
```

##  Calculating the Z score

```python
df_final['zscore_views'] = (df_final.viewcount - df_final.viewcount.mean())/df_final.viewcount.std(ddof=0)
df_final['zscore_like'] = (df_final.likeCount - df_final.likeCount.mean())/df_final.likeCount.std(ddof=0)
```

```python
df_final['score']=df_final['zscore_like']+df_final['viewcount']
del df_final['zscore_like']
del df_final['zscore_views']
```
